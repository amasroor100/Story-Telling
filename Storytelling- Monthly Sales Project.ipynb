```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales=pd.read_csv("month_sales.csv")
```

```python
sales
```

```python
fig,(ax1, ax2, ax3, ax4)= plt.subplots(4,1, figsize=(6,8))
axes = [ax1, ax2, ax3, ax4]
for i in axes:
    i.plot(sales["Months"], sales["sales"], alpha=0.7)
   # Removing Labels
    i.set_yticklabels([])
    i.set_xticklabels([])
    
    # Removing Ticks
    i.tick_params(bottom=0, left=0)
    
    # Removing Borders
#     i.spines["left"].set_visible(False)
    loc=["left", "right", "top", "bottom"]
    
    for j in loc:
        i.spines[j].set_visible(False)

        # Plotting Quaterly   
ax1.plot(sales["Months"][0:3], sales["sales"][0:3], linewidth=2.5, color= "Blue")
ax2.plot(sales["Months"][2:6], sales["sales"][2:6], linewidth=2.5, color= "Blue")
ax3.plot(sales["Months"][5:9], sales["sales"][5:9], linewidth=2.5, color= "Blue")
ax4.plot(sales["Months"][8:12], sales["sales"][8:12], linewidth=2.5, color= "Blue")
    # label
ax1.text(0.5, -80, "0", alpha=0.5) 
ax1.text(2.2, 58000, "55691.01", alpha=0.5) 
    
ax2.text(2.2, 58000, "55691.01", alpha=0.5) 
ax2.text(5.5, 37000, "34595.13", alpha=0.5) 

ax3.text(5.5, 37000, "34595.13", alpha=0.5) 
ax3.text(8.3, 84000, "81777.35", alpha=0.5)

ax4.text(8.3, 84000, "81777.35", alpha=0.5) 
ax4.text(11, 61000, "69545.62", alpha=0.5)

ax1.text(2.4, 15000, "Jan-Mar", weight='bold', rotation=65, color="Blue")
ax2.text(3.4, 13000, "Apr-Jun", weight='bold', rotation=-10, color="Blue")
ax3.text(7.6, 40000, "Jul-Sep", weight='bold', rotation=70, color="Blue")
ax4.text(10.1, 55000, "Oct-Dec", weight='bold', rotation=65, color="Blue")

ax1.text(0.5, 130000, "Quarterly Performance Review of Market", size=14, weight="bold")
ax1.text(0.5, 100000, "Total Sales in each quater", size=11)


plt.show()
```
