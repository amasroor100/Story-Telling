```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
crash = pd.read_csv("1938_airplane_crashes.csv", header=1)
```

```python
crash
```

```python
fig , (ax1,ax2) = plt.subplots(2,1 , figsize = (14,10))


ax1.bar(crash["Months"], crash["Aboard"], label='Aboard', color='Gold')
ax2.bar(crash["Months"], crash["Fatalities"], label='Fatalities', color='aqua')


# ax.tick_params(bottom=False, left=True)
ax1.legend(frameon=False, fontsize=15)
ax2.legend(frameon=False, fontsize=15)

   # Removing Labels
ax1.set_yticklabels([])
ax1.set_xticklabels(crash['Months'], rotation=35, fontsize=8)
ax2.set_yticklabels([])
ax2.set_xticklabels(crash['Months'], rotation=35, fontsize=8)    
#     # Removing Ticks
ax1.tick_params(bottom=0, left=0)
ax2.tick_params(bottom=0, left=0)   
    # Removing Borders
#     i.spines["left"].set_visible(False)
loc=["left", "right", "top", "bottom"]
    
for j in loc:
    ax1.spines[j].set_visible(False)
    ax2.spines[j].set_visible(False)
for i in range(len(crash["Months"])):
    ax1.text(i-0.2,crash["Aboard"][i]+0.2,crash["Aboard"][i],weight='bold', fontsize=10)
    ax2.text(i-0.2,crash["Fatalities"][i]+0.2,crash["Fatalities"][i],weight='bold', fontsize=10)

ax1.text(0.2, 20, "Monthly Representation of Aboard Passengers in Airplance Crashes in 1938", weight='bold', fontsize=16)
ax2.text(0.2, 20, "Monthly Representation of Fatalities in Airplance Crashes in 1938", weight='bold', fontsize=16)
```

```python
fig, ax = plt.subplots(figsize=(16,7))


ax.bar(crash["Months"], crash["Aboard"], label='Aboard', color='Gold')
ax.bar(crash["Months"], crash["Fatalities"], label='Fatalities', color='aqua')


# ax.tick_params(bottom=False, left=True)
ax.legend(frameon=False, fontsize=17)


   # Removing Labels
ax.set_yticklabels([])
ax.set_xticklabels(crash['Months'], rotation=45, fontsize=12)
    
#     # Removing Ticks
ax.tick_params(bottom=0, left=0)
    
    # Removing Borders
#     i.spines["left"].set_visible(False)
loc=["left", "right", "top", "bottom"]
    
for j in loc:
    ax.spines[j].set_visible(False)

for i in range(len(crash["Months"])):
    ax.text(i-0.2,crash["Aboard"][i]+0.5,crash["Aboard"][i],weight='bold', fontsize=16)

for i in range(len(crash["Months"])):
    if (crash["Aboard"][i]!=0):
        ax.annotate((crash['Location'][i]), xy=(i, (crash['Aboard'][i]/2)),\
                    ha='center', va='center', weight='bold', size=8, rotation=80)
    else:
        ax.annotate((crash['Location'][i]), xy=(i, (crash['Aboard'][i]+3)),\
                    ha='center', va='center', weight='bold', size=8, rotation=80)
    

ax.text(0.2, 25, "Monthly Representation of Airplance Crashes in 1938", weight='bold', fontsize=25)
ax.text(0, 23,"Total People Traveled in 1938 on Crashed Airplane are: " +str(crash["Aboard"].cumsum()[11]), fontsize=16)
ax.text(0, 21,"Total Fatalities due to Airplance Crash in 1938 are: " +str(crash["Fatalities"].cumsum()[11]), fontsize=16)
```
